<span style="font-size:30px; font-family:Courier;"> Creating a directory and file </span>

```python
from pathlib import Path
p = Path.cwd() / 'spam'
p.mkdir(exist_ok=True)
with open(p / 'file1.txt', 'w', encoding='utf-8') as file:
    file.write("Hello Python")
```

<span style="font-size:30px; font-family:Courier;"> Copy using shutil </span>


```python
from pathlib import Path
import shutil
dest = Path.cwd() / 'dest'
src = Path.cwd() / 'spam' / 'file1.txt'
srcd = Path.cwd() / 'spam'
dest.mkdir(exist_ok=True)
shutil.copy(src, dest)
shutil.move(srcd/'file1.txt', dest /'file2.txt')
```

<span style="font-size:30px; font-family:Courier;"> Walking a Directory Tree </span>

```python
import os
from pathlib import Path
os.listdir('/home/koba/gitrepos')
githome = Path('/home/koba/gitrepos')
for folder_name, subfolders, filenames in os.walk('/home/koba/gitrepos'):
    print('Current folder is ' + folder_name)
    for subfolder in subfolders:
        print('Subfolder inside ' + folder_name + ' is ' + subfolder)
    for filename in filenames:
        print('File name inside ' + folder_name + ' is ' + filename )
    
list(githome.iterdir())
```
